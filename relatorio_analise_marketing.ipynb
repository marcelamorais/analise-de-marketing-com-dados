# ðŸ“Œ CÃ³digo:

```python
# Importar bibliotecas:

import pandas as pd
import json
```

```python
# Carregar arquivo JSON:

df = pd.read_json('/content/dados_mkt_campanhas.json')
```

```python
# Converter data para datetime:

df['data'] = pd.to_datetime(df['data'])
```

```python
# Criar coluna de trimestre:

df['trimestre'] = df['data'].dt.to_period('Q').astype(str)
```

```python
# Agrupar por canal e trimestre:

resumo = (
    df.groupby(['trimestre', 'canal'])
      .agg(
          investimento_total=('gasto', 'sum'),
          cliques_totais=('cliques', 'sum')
      )
      .reset_index()
)

print(resumo)
```

```python
# Calcular CPC (custo por clique):

resumo['cpc'] = resumo['investimento_total'] / resumo['cliques_totais']
print(resumo['cpc'])
```

```python
# Ajustar para 3 casas decimais:

resumo[['investimento_total', 'cliques_totais', 'cpc']] = resumo[
    ['investimento_total', 'cliques_totais', 'cpc']
].round(3)

print(resumo[['investimento_total', 'cliques_totais', 'cpc']])
```

```python
# Ordenar para facilitar anÃ¡lise:

resumo = resumo.sort_values(['trimestre', 'investimento_total'], ascending=[True, False])
print(resumo)
```
